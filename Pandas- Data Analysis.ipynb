# Chapter 5- Apply

```python
student_details = pd.read_csv('student-mat.csv')
student_details
```

```python
student_details['Medu']
```

```python
student_details['Medu'] = student_details['Medu'] + 2
```

```python
student_details['Mjob'] = student_details['Mjob'].str.upper() #convert into str and then upper
```

```python
student_details
```

```python
student_details['Fjob'] = student_details['Fjob'].apply(lambda x:x.upper()) #lambda is an anonymous function
                                                  #While normal functions are defined using def, keys are defined using lambda keyword.
                                                  #Hence, anonymous functions are also called lamba functions.
                                                  #It takes only one expression.
```

```python
student_details
```
