# Task -5 : Data Cleaning

```python
import pandas as pd
```

## 1. Read the following .csv files:
1. Sales Transactions-2017.csv
2. Sales Transactions-2018.csv
3. Sales Transactions-2019.csv

```python
sales_2017 = pd.read_csv('Sales Transactions-2017.csv')
sales_2017
```

```python
sales_2018 = pd.read_csv('Sales Transactions-2018.csv')
sales_2018
```

```python
sales_2019 = pd.read_csv('Sales Transactions-2019.csv')
sales_2019
```

## 3. Concatenate the 3 files:

```python
complete_data = sales_2017.append([sales_2018, sales_2019])
```

```python
complete_data
```

## 4. Deletion of null dates from Date feature in dataframe:

```python
complete_data.dropna(subset=['Date'], inplace=True)
```

```python
complete_data
```

## 5. Changing object Date into datetime datatype

```python
complete_data['Date'] = pd.to_datetime(complete_data['Date'], errors='coerce')
```

```python
complete_data
```

```python
complete_data.info()
```

## 6. Deletion of unrequired features(columns):

```python
filtered_data = complete_data.drop(labels = ['Disc', 'Voucher Amount', 'Gross'], axis= 1)
```

```python
filtered_data
```

## 7. Checking for null values:

```python
filtered_data.isnull().sum()
```

```python
import warnings 
warnings.filterwarnings(action='ignore')
```

## 8. Replacing characters in feature objects:

```python
filtered_data[['Qty', 'Rate']] = filtered_data[['Qty','Rate']].replace(',', '', regex=True)
```

```python
filtered_data.head(30)
```

## 9. Converting objects in features: Qty, Rate to numeric values and respective datatypes:

```python
filtered_data['Rate'] = pd.to_numeric(filtered_data['Rate']).astype('float')
```

```python
filtered_data['Qty'] = pd.to_numeric(filtered_data['Qty']).astype('int')
```

```python
filtered_data.dtypes
```

```python
filtered_data['Voucher'] = filtered_data['Voucher'].str.replace('Sal:','').astype('int')
```

```python
filtered_data
```

```python
filtered_data.info()
```

## 10. Checking for the position of present null value and filling it.:

```python
filtered_data[filtered_data['Rate'].isnull()].index.tolist()
#df[df['A'].isnull()].index.tolist()
```

```python
filtered_data['Rate'] = filtered_data['Rate'].fillna(0)
```
